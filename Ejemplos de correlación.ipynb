## Correlación  perfectamente negativa

```python
import matplotlib.pyplot as plt
from pydataset import data
import pandas as pd
```

```python
#función para el coeficiente de correlación
from math import sqrt
def cof_corr(Ac_A,Ac_B):
    mean_A=sum(Ac_A)/len(Ac_A)
    mean_B=sum(Ac_B)/len(Ac_B)
    cov=sum([(a-mean_A)*(b-mean_B) for a,b in zip(Ac_A,Ac_B)])
    sigma_A=sqrt(sum([(a-mean_A)**2 for a in Ac_A]))
    sigma_B=sqrt(sum([(b-mean_B)**2 for b in Ac_B]))
    p=cov/(sigma_A*sigma_B)
    return p
```

```python
#Importando el comportamiendo del SP500
SP500=(1+data("SP500")).cumprod()
```

## Correlación perfectamente positiva

```python
#DATOS
Dic_pos={'ACTIVO A':SP500["r500"],'ACTIVO B':0.5*SP500["r500"]}
Dat_pos=pd.DataFrame(Dic_pos)
Dat_pos
```

```python
#COEFICINETE DE CORRELACIÓN 
cof_corr(Dat_pos['ACTIVO A'],Dat_pos['ACTIVO B'])
```

```python
#GRAFICA
plt.plot(Dat_pos['ACTIVO A'],label="Activo A")
plt.plot(Dat_pos['ACTIVO B'],label="Activo B")
plt.title("r=1")
plt.legend(loc="upper left")
plt.show()
```

## Correlación perfectamente negativa 

```python
#Datos
Dic_neg={'Activo A':SP500["r500"],'Activo B':-1*SP500["r500"]}
Dat_neg=pd.DataFrame(Dic_neg)
Dat_neg
```

```python
cof_corr(Dat_neg["Activo A"],Dat_neg["Activo B"])
```

```python
plt.plot(Dat_neg["Activo A"],label="Activo A")
plt.plot(Dat_neg["Activo B"],label="Activo B")
plt.title("r=-1")
plt.legend(loc="lower left")
plt.show()
```

## Coeficiente correlación nulo

```python
#Datos
x=[6,8,6,8]
y=[1,2,2,1]
```
