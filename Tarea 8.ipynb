```python
#Función para crear los gradficos de las regresiones
def grafico_regresion(A,B,col1="blue",col2="red"):
    x,y = np.array(Datos[A]),Datos[B]
    regresion_lineal = LinearRegression() 
    regresion_lineal.fit(x.reshape(-1,1), y) 
    b0,b1=regresion_lineal.coef_[0],regresion_lineal.intercept_
    plt.title(A+" VS "+B)
    plt.scatter(x,y,color=col1)
    a=np.linspace(-0.08,0.06,50)
    recta=f'Y={round(regresion_lineal.coef_[0],5)}X+{round(regresion_lineal.intercept_,5)}'
    plt.plot(a,a*b0+b1,label=recta,color=col2)
    plt.legend(loc="upper left")
    plt.show()
```

#Grafico de las regresiones de IPC

```python
for col in Datos.columns[3:]:
    grafico_regresion("IPC",col)
```

#Grafico de las regresiones con la cartera T

```python
for col in Datos.columns[3:]:
    grafico_regresion("CarteraT",col)
```

#Cálculo de betas

```python
#BETAS
Bi=lambda A,B:np.cov(Datos[A],Datos[B])[0,1]/Datos[A].var()
IpcBi=[Bi("IPC",col)  for col in Datos.columns[1:] if col!='IPC' and col!='CarteraT' ]
CarTBi=[Bi('CarteraT',col)  for col in Datos.columns[1:] if col!='CarteraT' and col!='IPC']
DataBetas=pd.DataFrame({'IPC':IpcBi,'CarteraT':CarTBi}).T
DataBetas=DataBetas.rename(columns={DataBetas.columns[i]:Datos.columns[3:][i] for i in DataBetas.columns})
DataBetas
```
